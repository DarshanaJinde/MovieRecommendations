```python
min_v = df['No_of_Votes'].quantile(.9)
min_v
```

```python
mean_r = df['IMDB_Rating'].mean()
mean_r
```

```python
max_v_movies = df.loc[df['No_of_Votes'] >= min_v]
max_v_movies.shape
```

```python
def simple_recommender(x , mean_r = mean_r , min_v = min_v):
    r = x['IMDB_Rating']
    v = x['No_of_Votes']
    return ( ( v / (v + min_v) * r ) + ( min_v / (v + min_v) * mean_r ) )
```

```python
max_v_movies["Popularity"] = max_v_movies.apply(simple_recommender, axis=1)
max_v_movies=max_v_movies.sort_values(by="Popularity",ascending=False, )
max_v_movies = max_v_movies.round({'weighted_avg': 2})
max_v_movies.head(50)
```

```python
df["Overview"]
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
df['Director'] = df['Director'].apply(collapse)
df['Star1'] = df['Star1'].apply(collapse)
df['Star2'] = df['Star2'].apply(collapse)
df['Star3'] = df['Star3'].apply(collapse)
df['Star4'] = df['Star4'].apply(collapse)
df['Overview'] = df['Overview'].apply(lambda x:x.split())
df['Genre'] = df['Genre'].apply(lambda x:x.split())
```

```python
df['Overview'].head()
```

```python
df['Genre'].head()
```

```python
df['tags'] = df['Overview'] + df['Genre'] + df['Director'] + df['Star1'] + df['Star2'] + df['Star3'] + df['Star4'] + df['Series_Title'].apply(lambda x:x.split())
new = df.drop(columns=['Director', 'Genre', 'Overview', 'Star1', 'Star2', 'Star3', 'Star4', 'Runtime', 'No_of_Votes'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags'])
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['Series_Title'] == 'The Lord of the Rings: The Return of the King'].index[0]
```

```python
new.shape
```

```python
def recommend(movie):
    index = new[new['Series_Title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Series_Title)
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])

for i in distance[0:5]:
    print(new.iloc[i[0]].Series_Title)
```

```python
recommend('The Lord of the Rings: The Return of the King')
```
